```python
import pandas as pd
import json
from pyvis.network import Network

# Load the JSON data
issues_data = pd.read_json('../flutter_all_issues.json')

# Create a Network object from PyVis
net = Network(height="100%", width="100%", bgcolor="#ffffff", font_color="black")

# Options for nodes and physics
options = {
    "nodes": {
        "size": 20,
        "font": {"size": 16},
        "shape": "dot"
    },
    "physics": {
        "barnesHut": {
            "gravitationalConstant": -80000,
            "centralGravity": 0.3,
            "springLength": 200
        },
        "minVelocity": 0.75,
    }
}

# Set the options
net.set_options(json.dumps(options))

# Dictionary to store labels and their corresponding issue numbers
label_dict = {}

# Populate the dictionary and add nodes
for _, issue in issues_data.iterrows():
    issue_number = issue['number']
    issue_labels = [label['name'] for label in issue['labels']]
    net.add_node(issue_number, label=str(issue_number), title=issue['title'])
    for label in issue_labels:
        if label not in label_dict:
            label_dict[label] = []
        label_dict[label].append(issue_number)

# Add edges based on shared labels
for label, issues in label_dict.items():
    for i in range(len(issues)):
        for j in range(i + 1, len(issues)):
            net.add_edge(issues[i], issues[j])

# Save the network diagram to an HTML file
output_file_path = 'flutter_issues_network_adjusted_all_issues.html'
net.save_graph(output_file_path)

# The output path for downloading or viewing the diagram
output_file_path
```
